## Discussion

Our initial expectation was that there would be a negative linear relationship between mortality and vaccination rates. However, in 2021 and 2022, there is no linear relationship between mortality and single/multiple vaccination rates. Limitations include data collection and collation, variable selection, and model selection.

Data collection issues are probably the least likely to lead to poor model fit. However, the most important constraint in this category is that we need to know how the data in the vaccination dataset was collected. We presume that it was scraped through surveys and the web. If the data were collected through surveys, then the people who answered the surveys are biased and there is no definitive way to prove whether the subjects are telling the truth. In this case, people have reasons to lie about vaccination status, such as not wanting to be vaccinated, but appearing to do so to avoid conflict. If data are obtained through web scraping, it is likely that data will be lost or inconsistent. Large-scale web scraping, such as vaccination status for all countries, can also run slowly, which may mean taking shortcuts to reduce the time it takes to scrape data.

Collation problems may be an important factor in the poor fit of our linear regression model. Since we merged two datasets, the first for COVID cases and deaths and the second for vaccinations, there may have been a loss of data due to insufficient cleaning and standardization of the data prior to merging. Vaccination rates appear to be cumulative in the merged dataset, so we attempted to make them monthly. In the collation, there may have been an error in counting rather than accumulating, or we may have misjudged the cumulative vaccination rates. However, this is very unlikely. It is also possible that human error in the calculations contributed to these models. In addition, variable selection may have been a limiting factor, as all rates were aggregated rather than from the original source. As a result, negative vaccination rates emerged that should not have been present.

A final possible limitation is that the linear regression model is not appropriate for our data set. Although linear regression performs well for predicting one-to-one relationships between the different values in this case, it is sensitive to outliers and noise. In such a complex analysis of a polarizing measure to prevent COVID-19, there would be outliers and other variables in mortality. This was especially true in the years when vaccines were first introduced to the public. Another drawback of linear regression is the assumption of linearity between the dependent and independent variables. As we can see from the graphs, there does not appear to be a linear relationship between both cases and years.

Nonetheless, the model fit is correct and these variables are not correlated. We know that the main goal of the vaccine is to prevent patients from being infected, which should mitigate the increase in mortality. Although a small reduction in severity has been claimed, there does not seem to be a definitive study on the reduction of the risk of death from COVID-19 infection. Another step that could complement this study would be to see if there is a linear relationship between vaccination rates and case rates. If we could find a linear relationship between these variables, this would likely mean that there should be a relationship between the vaccine and COVID-19 mortality, since it is not possible to die from COVID-19 without first contracting the virus.

Finally, it is worth noting that the data are for 2021 and 2022 only. The vaccine is first introduced to the public in 2021 and is not available to all age groups until 2022. This means that we will only know the true efficacy of the vaccine after more time has passed. Time constraints prevent us from seeing the full picture, including our study and other studies on COVID-19. Also, after dropping states with lots of missing datas, our conclusion is limited to the states included in our sample.

To improve our study, we need to look for more reliable data and get access to a complete data information. Moreover, we can look for more dataset and merge potential variables to create a more comprehensive dataset. When dealing with missing data, we need to learn more method such as impute it from the other rows, instead of dropping them directly. Last but not least, model selection is also essential. When we fail to see a significant linear trend, we can consider fitting more advanced model and see our result.
